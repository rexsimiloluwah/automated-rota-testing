```python
# Parse pytest output into rows
def parse_pytest(output):
    rows = [["Test File", "Test Class", "Test Name", "Status"]]
    for line in output.strip().splitlines():
        match = re.match(r"(\S+)::(\S+)::(\S+)\s+(PASSED|FAILED)", line)
        if match:
            status = "вњ… PASSED" if match.group(4) == "PASSED" else "вќЊ FAILED"
            rows.append([match.group(1), match.group(2), match.group(3), status])
    return rows

# Create spreadsheet
sh = gc.create(SPREADSHEET_NAME)

# Move to folder
drive = build('drive', 'v3', credentials=creds)
drive.files().update(
    fileId=sh.id,
    addParents=FOLDER_ID,
    removeParents='root',
    fields='id, parents'
).execute()

# Helper: style a sheet
def style_sheet(worksheet, num_rows, num_cols):
    """Apply header styling, text wrap, and column widths."""
    # Bold white text on dark blue header
    worksheet.format("1:1", {
        "textFormat": {"bold": True, "fontSize": 11, "foregroundColorStyle": {"rgbColor": {"red": 1, "green": 1, "blue": 1}}},
        "backgroundColor": {"red": 0.2, "green": 0.3, "blue": 0.5},
        "horizontalAlignment": "CENTER",
        "wrapStrategy": "WRAP",
    })

    # Wrap all data cells and set font size
    if num_rows > 1:
        worksheet.format(f"2:{num_rows}", {
            "wrapStrategy": "WRAP",
            "textFormat": {"fontSize": 10},
            "verticalAlignment": "TOP",
        })

    # Auto-resize columns
    sheet_id = worksheet._properties["sheetId"]
    requests = []
    col_widths = {0: 300, 1: 350, 2: 200, 3: 150}
    for col_idx in range(num_cols):
        width = col_widths.get(col_idx, 200)
        requests.append({
            "updateDimensionProperties": {
                "range": {
                    "sheetId": sheet_id,
                    "dimension": "COLUMNS",
                    "startIndex": col_idx,
                    "endIndex": col_idx + 1,
                },
                "properties": {"pixelSize": width},
                "fields": "pixelSize",
            }
        })

    # Set row height for data rows
    if num_rows > 1:
        requests.append({
            "updateDimensionProperties": {
                "range": {
                    "sheetId": sheet_id,
                    "dimension": "ROWS",
                    "startIndex": 1,
                    "endIndex": num_rows,
                },
                "properties": {"pixelSize": 35},
                "fields": "pixelSize",
            }
        })

    # Header row height
    requests.append({
        "updateDimensionProperties": {
            "range": {
                "sheetId": sheet_id,
                "dimension": "ROWS",
                "startIndex": 0,
                "endIndex": 1,
            },
            "properties": {"pixelSize": 40},
            "fields": "pixelSize",
        }
    })

    # Freeze header row
    requests.append({
        "updateSheetProperties": {
            "properties": {
                "sheetId": sheet_id,
                "gridProperties": {"frozenRowCount": 1},
            },
            "fields": "gridProperties.frozenRowCount",
        }
    })

    sh.batch_update({"requests": requests})


# Sheet 1: Unit Tests
ws1 = sh.sheet1
ws1.update_title("Unit Tests")
unit_rows = parse_pytest(pytest_output)
ws1.update(f"A1:D{len(unit_rows)}", unit_rows)
style_sheet(ws1, len(unit_rows), 4)

# Sheet 2: Notebook Checks
ws2 = sh.add_worksheet(title="Notebook Checks", rows=40, cols=4)
nb_rows = [["Status", "Notebook", "Course", "Details"]]
for status, name, details in notebook_results:
    course = name.split("_lab_")[0].replace("gdm_", "") if "_lab_" in name else ""
    if status == "PASS":
        emoji_status = "вњ… PASS"
    elif status == "FAIL":
        emoji_status = "вќЊ FAIL"
    else:
        emoji_status = "вЏ­пёЏ SKIP"
    nb_rows.append([emoji_status, name, course, details])
ws2.update(f"A1:D{len(nb_rows)}", nb_rows)
style_sheet(ws2, len(nb_rows), 4)

# Sheet 3: Smoke Tests 
ws3 = sh.add_worksheet(title="Smoke Tests", rows=40, cols=4)
smoke_rows = parse_pytest(pytest_output)
ws3.update(f"A1:D{len(smoke_rows)}", smoke_rows)
style_sheet(ws3, len(smoke_rows), 4)

# Sheet 4: GPU Tests (placeholder)
ws4 = sh.add_worksheet(title="GPU Tests", rows=5, cols=3)
ws4.update("A1:C2", [
    ["Status", "Notebook", "Details"],
    ["вЏёпёЏ DISABLED", "GPU tests not yet enabled", "Pending WIF setup"],
])
style_sheet(ws4, 2, 3)

# Sheet 5: Summary
ws5 = sh.add_worksheet(title="Summary", rows=10, cols=3)
passed = sum(1 for s, _, _ in notebook_results if s == "PASS")
skipped = sum(1 for s, _, _ in notebook_results if s == "SKIP")
failed = sum(1 for s, _, _ in notebook_results if s == "FAIL")

summary_rows = [
    ["Metric", "Value", "Notes"],
    ["Run Date", datetime.now().strftime("%Y-%m-%d %H:%M:%S"), ""],
    ["Python Version", "3.12.13", "Colab runtime"],
    ["Unit Tests", "вњ… 28 passed", ""],
    ["Notebooks Passed", f"вњ… {passed}", ""],
    ["Notebooks Skipped", f"вЏ­пёЏ {skipped}", "GPU required"],
    ["Notebooks Failed", f"{'вќЊ' if failed else 'вњ…'} {failed}", ""],
]
ws5.update(f"A1:C{len(summary_rows)}", summary_rows)
style_sheet(ws5, len(summary_rows), 3)

print(f"Spreadsheet created: {sh.url}")
```
